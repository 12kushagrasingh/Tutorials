# Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
#creating sample data of 100 students and specific streams in class
classes = ["Python", "Data Science", "DL", "ML", "AI"]
class1 = [30, 10, 20, 25, 10]
class2 = [40, 5, 20, 20, 10]
class3 = [35, 5, 30, 15, 15]
```

```python
#plotting bar chart, deafult
plt.bar(classes, class1)
```

```python
#plotting bar chart horizontally
plt.barh(classes, class1)
```

# width()

```python
#width of bar
plt.bar(classes, class1, width = 0.9) #0.9 means 90%

#try yourself: horizontal bar with width parameter
#width is not possible in horizontal bar, it will be multiple arguments
```

# align()

```python
#align : {'center', 'edge'}, default: 'center', alignment of bar
plt.bar(classes, class1, width = 0.7, align = 'edge')
```

# color()

```python
#changing color of bar
plt.bar(classes, class1, width = 0.7, align = 'edge', color='y')
```

# edgecolor()

```python
#changing border color of the bar
plt.bar(classes, class1, width = 0.7, align = 'edge', color='y', edgecolor='r')
```

# linewidth()

```python
#changing width of borders
plt.bar(classes, class1, width = 0.7, align = 'edge', color='y', edgecolor='r', linewidth = 3)
```

# aplha()

```python
#changing transparency of color
plt.bar(classes, class1, width = 0.7, align = 'edge', color='y', edgecolor='r', linewidth = 3, alpha = 0.6)
```

# linestyle()

```python
#changing style of border line
plt.bar(classes, class1, width = 0.7, align = 'edge', color='y', edgecolor='r', linewidth = 3, alpha = 0.9, linestyle = ':')
```

# visible()

```python
#changing the visiblity of the bar chart
plt.bar(classes, class1, width = 0.7, align = 'edge', color='y', edgecolor='r', linewidth = 3, alpha = 0.9, linestyle = ':', visible = False)
```

# ggplot

```python
#providing grid to bar chart
style.use('ggplot')
plt.bar(classes, class1, width = 0.7, align = 'edge', color='k', edgecolor='r', linewidth = 3)
```

# graph size

```python
#Changing the size of barchart
style.use('ggplot')
plt.figure(figsize = (16,9))
plt.bar(classes, class1, width = 0.7, align = 'edge', color='y', edgecolor='r', linewidth = 3)
```

# title and Fontsize

```python
#providing titles and setting font size
style.use('ggplot')
plt.bar(classes, class1, width = 0.6, color='k', edgecolor='r', linewidth = 3)
plt.title("Strength", fontsize = 21) #this gives title to graph
plt.xlabel('Class', fontsize = 12) #gives label to x axis
plt.ylabel('No of Students', fontsize = 12) #gives label to y axis
plt.show() #this shows the data to display
```

# Multiple Bar Charts

```python

style.use('ggplot')

classes = np.arange(len(classes)) #refer numpy tutorial how arrange function works
width = 0.3

plt.bar(classes, class1, width, color='k', edgecolor='r', linewidth = 3)
plt.bar(classes + width, class2, width, color='y', edgecolor='r', linewidth = 3) #by adding width we are changing the location

plt.title("Strength", fontsize = 21) #this gives title to graph
plt.xlabel('Class', fontsize = 12) #gives label to x axis
plt.ylabel('No of Students', fontsize = 12) #gives label to y axis
plt.show() #this shows the data to display
```

```python

style.use('ggplot')

plt.figure(figsize = (16,9))

classes = np.arange(len(classes)) #refer numpy tutorial how arrange function works
width = 0.3
plt.bar(classes, class1, width, color='g', edgecolor='r', linewidth = 3)
plt.bar(classes + width, class2, width, color='b', edgecolor='r', linewidth = 3) #by adding width we are changing the location
plt.bar(classes+width+width, class3, width, color='y', edgecolor='r', linewidth = 3)

plt.xticks(classes+width, classes, rotation = 30) #"classes + width" title is changed with classes

plt.title("Strength", fontsize = 21) #this gives title to graph
plt.xlabel('Class', fontsize = 12) #gives label to x axis
plt.ylabel('No of Students', fontsize = 12) #gives label to y axis
plt.legend(['Class 1', 'Class 2', 'Class 3'], loc = 1) #gives name to the line, loc is used to set the location of label
plt.show() #this shows the data to display
```
