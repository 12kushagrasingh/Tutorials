# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
#creating a sample data using random
ml_student_age = np.random.randint(18,42, (100))
py_student_age = np.random.randint(15,32, (100))

print(ml_student_age)
print(py_student_age)
```

# hist()

```python
plt.hist(ml_student_age) #hist is used to given for which histogram will be plotted
plt.title("Age Histogram") #this gives title to graph
plt.xlabel('Age Category') #gives label to x axis
plt.ylabel('No of Students') #gives label to y axis
plt.show() #this shows the data to display
```

### bins()

```python
#bins make the bars between the defined limits
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins) #hist is used to given for which histogram will be plotted
plt.title("Age Histogram") #this gives title to graph
plt.xlabel('Age Category') #gives label to x axis
plt.ylabel('No of Students') #gives label to y axis
plt.show() #this shows the data to display
```

### rwidth()

```python
#rwidth will give the width between the bars
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins, rwidth = 0.6) #0.6 means 60%
plt.title("Age Histogram") #this gives title to graph
plt.xlabel('Age Category') #gives label to x axis
plt.ylabel('No of Students') #gives label to y axis
plt.show() #this shows the data to display
```

### type()

```python
#histtype : {'bar', 'barstacked', 'step',  'stepfilled'}, these are type of histogram to draw.
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins, rwidth = 0.6, histtype = 'step') #0.6 means 60%
plt.title("Age Histogram") #this gives title to graph
plt.xlabel('Age Category') #gives label to x axis
plt.ylabel('No of Students') #gives label to y axis
plt.show() #this shows the data to display

#Try yourslef: Other types
```

### align()

```python
#align : {'left', 'mid', 'right'}, use to align bar
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins, rwidth = 0.6, align='left') #0.6 means 60%
plt.title("Age Histogram") #this gives title to graph
plt.xlabel('Age Category') #gives label to x axis
plt.ylabel('No of Students') #gives label to y axis
plt.show() #this shows the data to display

#Try yourslef: Other align
```

### orientation()

```python
# orientation : {'horizontal', 'vertical'}, decides the orientation
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins, rwidth = 0.6, orientation='horizontal') #0.6 means 60%
plt.title("Age Histogram") #this gives title to graph
plt.xlabel('Age Category') #gives label to x axis
plt.ylabel('No of Students') #gives label to y axis
plt.show() #this shows the data to display
```

### color()

```python
#use to change the color of histogram
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins, rwidth = 0.6, orientation='horizontal', color='m') #0.6 means 60%
plt.title("Age Histogram") #this gives title to graph
plt.xlabel('Age Category') #gives label to x axis
plt.ylabel('No of Students') #gives label to y axis
plt.show() #this shows the data to display
```

### legend()

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.hist(ml_student_age, bins, rwidth = 0.6, orientation='horizontal', color='m') #0.6 means 60%
plt.title("Age Histogram") #this gives title to graph
plt.xlabel('Age Category') #gives label to x axis
plt.ylabel('No of Students') #gives label to y axis
plt.legend(['Age'], loc = 1) #gives name to the line, loc is used to set the location of label
plt.show() #this shows the data ot display

#Try yourself: label can also be assigned using following code

# bins = [15, 20, 25, 30, 35, 40, 45]
# plt.hist(ml_student_age, bins, rwidth = 0.6, orientation='horizontal', color='m', label = "Age") #0.6 means 60%
# plt.title("Age Histogram") #this gives title to graph
# plt.xlabel('Age Category') #gives label to x axis
# plt.ylabel('No of Students') #gives label to y axis
# plt.legend(loc = 1) #gives name to the line, loc is used to set the location of label
# plt.show() #this shows the data ot display

```

# Multiple histogram

```python
bins = [15, 20, 25, 30, 35, 40, 45]

plt.hist(ml_student_age, bins, rwidth = 0.6, orientation='horizontal', color='g')
plt.hist(py_student_age, bins, rwidth = 0.6, orientation='horizontal', color='r')

plt.title("Age Histogram") #this gives title to graph
plt.xlabel('Age Category') #gives label to x axis
plt.ylabel('No of Students') #gives label to y axis
plt.legend(['Age py students', 'Age ml students'], loc = 1) #gives name to the line, loc is used to set the location of label
plt.show() #this shows the data ot display

#Graph is not proper and overlaped
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]

plt.hist(ml_student_age, bins, rwidth = 0.6, orientation='horizontal', color='g')
plt.hist(py_student_age, bins, rwidth = 0.6, orientation='vertical', color='r')

plt.title("Age Histogram") #this gives title to graph
plt.xlabel('Age Category') #gives label to x axis
plt.ylabel('No of Students') #gives label to y axis
plt.legend(['Age py students', 'Age ml students'], loc = 1) #gives name to the line, loc is used to set the location of label
plt.show() #this shows the data ot display

#Graph is not proper and overlaped
```

```python
bins = [15, 20, 25, 30, 35, 40, 45]

plt.hist([ml_student_age, py_student_age], bins, rwidth = 0.6, orientation='vertical', color=['y', 'm'])

plt.title("Age Histogram") #this gives title to graph
plt.xlabel('Age Category') #gives label to x axis
plt.ylabel('No of Students') #gives label to y axis
plt.legend(['Age py students', 'Age ml students'], loc = 1) #gives name to the line, loc is used to set the location of label
plt.show() #this shows the data ot display
```

# histogram size

```python
bins = [15, 20, 25, 30, 35, 40, 45]

plt.figure(figsize = (15,9)) #this will change the size of histogram

plt.hist([ml_student_age, py_student_age], bins, rwidth = 0.6, orientation='vertical', color=['y', 'm'])
plt.title("Age Histogram") #this gives title to graph
plt.xlabel('Age Category') #gives label to x axis
plt.ylabel('No of Students') #gives label to y axis
plt.legend(['Age py students', 'Age ml students'], loc = 1) #gives name to the line, loc is used to set the location of label
plt.show() #this shows the data ot display
```

# background and grid

```python
bins = [15, 20, 25, 30, 35, 40, 45]
plt.figure(figsize = (15,9)) #this will change the size of histogram

plt.grid(color = 'g', linestyle = '-', linewidth = 2) #gives grid to the background
#but the grid here is above histogram, so we have to use style module

plt.hist([ml_student_age, py_student_age], bins, rwidth = 0.6, orientation='vertical', color=['y', 'm'])
plt.title("Age Histogram") #this gives title to graph
plt.xlabel('Age Category') #gives label to x axis
plt.ylabel('No of Students') #gives label to y axis
plt.legend(['Age py students', 'Age ml students'], loc = 1) #gives name to the line, loc is used to set the location of label
plt.show() #this shows the data ot display
```
