# Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
#creating sample data of 100 students and specific streams in class
classes = ["Python", "Data Science", "DL", "ML", "AI"]
class1 = [30, 10, 20, 25, 10]
class2 = [40, 5, 20, 20, 10]
class3 = [35, 5, 30, 15, 15]
```

```python
#plotting pie chart single valued
plt.pie([1])
plt.show()
```

```python
#plotting pie chart single valued
plt.pie(class1)
plt.show()
```

```python
#plotting pie chart single valued
plt.pie(class1, labels = classes)
plt.show()
```

```python
#cut a specific peice of pie out
classes = ["Python", "Data Science", "DL", "ML", "AI"]
class1 = [30, 10, 20, 25, 10]

e = [0.2, 0, 0, 0.1, 0] #python came out by 20% and ML by 10%

plt.pie(class1, labels = classes, explode = e)
plt.show()
```

```python
#cut a specific peice of pie out
classes = ["Python", "Data Science", "DL", "ML", "AI"]
class1 = [30, 10, 20, 25, 10]

e = [0.2, 0, 0, 0.1, 0] #python came out by 20% and ML by 10%
c = ['r', 'g', 'b', 'y', 'c']

plt.pie(class1, labels = classes, explode = e, colors = c)
plt.show()
```

```python
#providing percentage of each share
classes = ["Python", "Data Science", "DL", "ML", "AI"]
class1 = [30, 10, 20, 25, 10]

e = [0.2, 0, 0, 0.1, 0] #python came out by 20% and ML by 10%
c = ['r', 'g', 'b', 'y', 'c']

plt.pie(class1, labels = classes, explode = e, colors = c, autopct = '%0.1f%%') #f means float, 0.1 means 1 value after decimal
plt.show()
```

```python
#Giving shadow to the pie
classes = ["Python", "Data Science", "DL", "ML", "AI"]
class1 = [30, 10, 20, 25, 10]

e = [0.2, 0, 0, 0.1, 0] #python came out by 20% and ML by 10%
c = ['r', 'g', 'b', 'y', 'c']

plt.pie(class1, labels = classes, explode = e, colors = c, autopct = '%0.1f%%', shadow = True)
plt.show()
```

```python
#changing radius of the pie
classes = ["Python", "Data Science", "DL", "ML", "AI"]
class1 = [30, 10, 20, 25, 10]

e = [0.2, 0, 0, 0.1, 0] #python came out by 20% and ML by 10%
c = ['r', 'g', 'b', 'y', 'c']

plt.pie(class1, labels = classes, explode = e, colors = c, autopct = '%0.1f%%', shadow = True, radius = 1.5)
plt.show()
```

```python
#changing angle of pie, python is first peice and will start at 90 degree (here)
classes = ["Python", "Data Science", "DL", "ML", "AI"]
class1 = [30, 10, 20, 25, 10]

e = [0.2, 0, 0, 0.1, 0] #python came out by 20% and ML by 10%
c = ['r', 'g', 'b', 'y', 'c']

plt.pie(class1, labels = classes, explode = e, colors = c, autopct = '%0.1f%%', shadow = True, radius = 1.5, startangle = 90)
plt.show()
```

```python
#changing angle of pie, python is first peice and will start at 90 degree (here)
classes = ["Python", "Data Science", "DL", "ML", "AI"]
class1 = [30, 10, 20, 25, 10]

e = [0.2, 0, 0, 0.1, 0] #python came out by 20% and ML by 10%
c = ['r', 'g', 'b', 'y', 'c']
t ={"fontsize":15}

plt.pie(class1, labels = classes, explode = e, colors = c, autopct = '%0.1f%%', shadow = True, radius = 1.5, textprops = t)
plt.show()
```

```python
#changing label distance from the center
classes = ["Python", "Data Science", "DL", "ML", "AI"]
class1 = [30, 10, 20, 25, 10]

e = [0.2, 0, 0, 0.1, 0] #python came out by 20% and ML by 10%
c = ['r', 'g', 'b', 'y', 'c']
t ={"fontsize":15}

plt.pie(class1, labels = classes, explode = e, colors = c, autopct = '%0.1f%%', shadow = True, radius = 1.5, textprops = t, 
        labeldistance = 1.2)
plt.show()
```

```python
#changing percentage distance from the center
classes = ["Python", "Data Science", "DL", "ML", "AI"]
class1 = [30, 10, 20, 25, 10]

e = [0.2, 0, 0, 0.1, 0] #python came out by 20% and ML by 10%
c = ['r', 'g', 'b', 'y', 'c']
t ={"fontsize":15}

plt.pie(class1, labels = classes, explode = e, colors = c, autopct = '%0.1f%%', shadow = True, radius = 1.5, textprops = t, 
        labeldistance = 1.2, pctdistance = 0.7)
plt.show()
```

```python
#rotating the labels
classes = ["Python", "Data Science", "DL", "ML", "AI"]
class1 = [30, 10, 20, 25, 10]

e = [0.2, 0, 0, 0.1, 0] #python came out by 20% and ML by 10%
c = ['r', 'g', 'b', 'y', 'c']
t ={"fontsize":15}

plt.pie(class1, labels = classes, explode = e, colors = c, autopct = '%0.1f%%', shadow = True, radius = 1.5, textprops = t, 
        labeldistance = 1.2, pctdistance = 0.7, rotatelabels = True)
plt.show()
```

```python
#creating wedge in the center
classes = ["Python", "Data Science", "DL", "ML", "AI"]
class1 = [30, 10, 20, 25, 10]

e = [0.2, 0, 0, 0.1, 0] #python came out by 20% and ML by 10%
c = ['r', 'g', 'b', 'y', 'c']
t ={"fontsize":15}
w = {'linewidth':3, 'width': 1, 'edgecolor':'k'}

plt.pie(class1, labels = classes, explode = e, colors = c, autopct = '%0.1f%%', shadow = True, radius = 1.5, textprops = t, 
        labeldistance = 1.2, pctdistance = 0.7, wedgeprops = w)
plt.show()

#Try yourself: Change the wedge values and observe the changes
```

```python
#changing percentage distance from the center
classes = ["Python", "Data Science", "DL", "ML", "AI"]
class1 = [30, 10, 20, 25, 10]

e = [0.2, 0, 0, 0.1, 0] #python came out by 20% and ML by 10%
c = ['r', 'g', 'b', 'y', 'c']
t ={"fontsize":15}

plt.pie(class1, labels = classes, explode = e, colors = c, autopct = '%0.1f%%', shadow = True, radius = 1.5, textprops = t, 
        labeldistance = 1.2, pctdistance = 0.7)
plt.legend(loc = 3)
plt.show()
```
