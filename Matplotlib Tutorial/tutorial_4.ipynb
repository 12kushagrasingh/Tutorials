# Scatter Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_covid = pd.read_csv("D://Sample data//india_covid.csv")
df_covid
```

```python
x = df_covid['death']
y = df_covid['cases']
```

# plotting scatter

```python
plt.scatter(x,y)
plt.title("India Covid", fontsize = 21) #this gives title to graph
plt.xlabel('Deaths', fontsize = 12) #gives label to x axis
plt.ylabel('Cases', fontsize = 12) #gives label to y axis
```

# changing color

```python
plt.scatter(x,y, c='m')
plt.title("India Covid", fontsize = 21) #this gives title to graph
plt.xlabel('Deaths', fontsize = 12) #gives label to x axis
plt.ylabel('Cases', fontsize = 12) #gives label to y axis
```

# Changing plot size

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c='m')
plt.title("India Covid", fontsize = 21) #this gives title to graph
plt.xlabel('Deaths', fontsize = 12) #gives label to x axis
plt.ylabel('Cases', fontsize = 12) #gives label to y axis
```

# Changing Marker

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c='m', marker = '*')
plt.title("India Covid", fontsize = 21) #this gives title to graph
plt.xlabel('Deaths', fontsize = 12) #gives label to x axis
plt.ylabel('Cases', fontsize = 12) #gives label to y axis
```

# Changing marker size

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c='m', marker = '*', s=90)
plt.title("India Covid", fontsize = 21) #this gives title to graph
plt.xlabel('Deaths', fontsize = 12) #gives label to x axis
plt.ylabel('Cases', fontsize = 12) #gives label to y axis
```

# Changing transparency

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c='m', marker = '*', s = 90, alpha = 0.6)
plt.title("India Covid", fontsize = 21) #this gives title to graph
plt.xlabel('Deaths', fontsize = 12) #gives label to x axis
plt.ylabel('Cases', fontsize = 12) #gives label to y axis
```

# Changing marker border color

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c='g', marker = '*', s = 90, alpha = 0.9, edgecolor='r')
plt.title("India Covid", fontsize = 21) #this gives title to graph
plt.xlabel('Deaths', fontsize = 12) #gives label to x axis
plt.ylabel('Cases', fontsize = 12) #gives label to y axis
```

# Changing marker border width

```python
plt.figure(figsize=(16,9))
plt.scatter(x,y, c='g', marker = '*', s = 120, alpha = 0.9, edgecolor='r', linewidth=3)
plt.title("India Covid", fontsize = 21) #this gives title to graph
plt.xlabel('Deaths', fontsize = 12) #gives label to x axis
plt.ylabel('Cases', fontsize = 12) #gives label to y axis
```
