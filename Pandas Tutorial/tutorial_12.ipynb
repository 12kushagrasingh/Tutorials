# loc()

```python
#access a group of rows and columns by labels or a boolean array
import pandas as pd
```

```python
df = pd.read_csv('D:\\Sample data\\pokemon.csv')
df
```

```python
df.loc[3] #all the labels of 3rd row are displayed
```

```python
df.loc[[3,6,9]] #values of row 3,6 and 9 are displayed
```

```python
df.loc[3, 'species'] #all the labels of 3rd row and species column are displayed
```

```python
df.loc[3:9, 'species'] #sliced row from 3 to 9 in cloumn species
```

```python
df.loc[df['speed'] > 140] #this gives all rows with speed above 140
```

```python
df.loc[df['speed'] > 140, ['species']] #this gives all rows with speed above 140 under cloumn species
```

# iloc()


```python
#this is integer location - based indexing
#accessing a group of rows and columns using integer
df.iloc[369]
```

```python
df.iloc[[369]]
```

```python
df.iloc[:, 2] #this gives all rows and 2nd column
```
