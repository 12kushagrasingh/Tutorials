# header()

```python
#this is used the change the header row in a csv file
df2 = pd.read_csv('D:\\Sample data\\google_row.csv', header = 1) #making 2nd row as header
df2
```

```python
#note: all rows above the assigned header row will be no tbe displayed
df3 = pd.read_csv('D:\\Sample data\\google_row.csv', header = 4) #making 3rd row as header
df3
```

```python
#now suppose by deafult there is no header (eg. in csv below) the header mentioned we want as data row, then
df4 = pd.read_csv('D:\\Sample data\\google_noheader.csv') 
df4
```

```python
#we'll use none to create no header, or deafult header to be exact
df5 = pd.read_csv('D:\\Sample data\\google_noheader.csv', header = None) 
df5
```

# prefix()

```python
# we can also modify the deafult header using prefix
df6 = pd.read_csv('D:\\Sample data\\google_noheader.csv', header = None, prefix = 'Column ') #we can use any prefix name instead of column 
#prefix can only be used with header
df6
```

# names()

```python
#we can also assign seprate names to the column
df7 = pd.read_csv('D:\\Sample data\\google_noheader.csv', names = ['a','b','c','d','e','f','g'])
#name will be assigned in the form of a list
df7
#try to assign less names than the desired columns present?
```
