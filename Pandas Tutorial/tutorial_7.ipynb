# na_values()

```python
#in type_2 column we can find values with NaN(Not a Number), that is no values are available there
#NaN can be of any data type

#We can also add different values as NaN
df2 = pd.read_csv('D:\\Sample data\\pokemon.csv', na_values = 'Yes')
df2
```

```python
#We can also add different values as NaN, for more than one we have to use the list
df3 = pd.read_csv('D:\\Sample data\\pokemon.csv', na_values = ['Yes', 'grass', 'fire', 'poison'])
df3
```
