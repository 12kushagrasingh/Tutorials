# fillna() Parameters

```python
#fills the NaN values with given values or input

import pandas as pd
```

```python
df = pd.read_csv('D:\\Sample data\\pokemon.csv')
df
```

# value

```python
df.fillna(0) #NaN is replaced with zero, like this we can use any scalar values
```

```python
df.fillna({'species':'UNknown', 'type_1':'INFINITY'}) #changing every column NaN values with different inputs, given with the help of a dictionary
```

# method

```python
df.fillna(method = 'ffill') #Fill NaN with the previous values            ffill stands for forward fill
#instead of ffill we can also use 'pad'
```

```python
df.fillna(method = 'bfill') #Fill NaN with the later values            ffill stands for backward fill
```

# axis

```python
# 0 stands for row or index and 1 stands column
# value or method needs to be specified with axis talked earlier here
df.fillna(method = 'ffill', axis = 1) #if axis not given by deafult axis is 0

#try bfill yourself
```

# limit

```python
#input is given to NaN values to decided rows or cloumn.
#like axis, limit also needs to be used with value or method
df.fillna(0, limit = 3) #value zero is apllied to row limit at 3
```

```python
df.fillna(method = 'pad', limit = 1 ) #forward fill(or pad) is apllied to row limit at 1
```

# inplace

```python
#it does not create a new data frame but modify the orginal data frame
#need to be used with value or method
#values to this are given in boolean
df.fillna(3, inplace = True) #all NaN values are given modified with 3
df
```
