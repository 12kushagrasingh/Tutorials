# dtype()

```python
#dtype is used to change the data type of of any column
df1 = pd.read_csv('D:\\Sample data\\pokemon.csv', dtype={'hp': 'float64', 'type_2': 'str'}) #can change multiple data type all at once
df1
```

# True() and False() values

```python
#values and string can be converted to true or false
df1 = pd.read_csv('D:\\Sample data\\pokemon.csv', true_values = ['Yes'])
df1
```

```python
#values and string can be converted to true or false
df1 = pd.read_csv('D:\\Sample data\\pokemon.csv', false_values = ['Yes'])
df1
```
