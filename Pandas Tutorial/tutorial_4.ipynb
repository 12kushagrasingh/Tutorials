# Writing in csv file - 1

```python
#importe pandas as pd
import pandas as pd
```

```python
#reading the csv file
df1 = pd.read_csv('D:\\Sample data\\google.csv')
df1
```

```python
#finding the type of file imported
type(df1)
```

```python
#finding all the columns present in the csv
df1.columns
```

```python
#accessing only one row
df2 = pd.read_csv('D:\\Sample data\\google.csv', nrows =1) # this function will access row from the start
df2
```

```python
#accessing 5 rows(from the start)
df3 = pd.read_csv('D:\\Sample data\\google.csv', nrows =5)
df3
```

```python
#accessing a column of the csv
df4 = pd.read_csv('D:\\Sample data\\google.csv', usecols=[3])
df4
```

```python
#accessing specific columns of the csv
df5 = pd.read_csv('D:\\Sample data\\google.csv', usecols=[0,1,4])
df5
```

```python
#csv with 1st row of no use
df6 = pd.read_csv('D:\\Sample data\\google_row.csv')
df6
```

```python
#skiping 1st row will make the 2nd row the index row 
df7 = pd.read_csv('D:\\Sample data\\google_row.csv', skiprows = 1)
df7
```

```python
#skiping a specific row in the csv
df8 = pd.read_csv('D:\\Sample data\\google_row.csv', skiprows = [3])
df8
```

```python
#skipping multiple specific rows
df9 = pd.read_csv('D:\\Sample data\\google_row.csv', skiprows = [1,3,6,9])
df9
```

```python
#replacing the index column with some column in csv
df1 = pd.read_csv('D:\\Sample data\\google.csv', index_col = 'Date')
df1
```

```python
#replacing the index column with some coulmn using it's index number
df1 = pd.read_csv('D:\\Sample data\\google.csv', index_col = 6)
df1
```
