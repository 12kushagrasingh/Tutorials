# replace() Parameters

```python
import pandas as pd
```

```python
df = pd.read_csv('D:\\Sample data\\pokemon.csv')
df
```

```python
#string replacement
df.replace('poison', 'Botulinum toxin') #poison string is replaced with Botulinum toxin(most toxic substance in world) string
```

```python
#integer replacement
df.replace(80, 123456789) # 80 is replaced with 123456789
```

```python
#replacing more than one value
df.replace([1,3,5,803,805,807,80], "HASH") #list values will be repalced by HASH

#Try it yourself: replace list of values with list
```

```python
#replacing specific column values
#we'll need to use industry
df.replace({"type_1": "fire", 'speed':80}, 'FIRE') #in list type_1 and speed, fire and 80 are replaced with Fire
```

# regex()

```python
#regex stands for regular expresion
df.replace('[a-z]', 0, regex = True) #all letters from a to z are replaced with zero
```

```python
#replacing strings in specific column
#we'll use dictionary for this
df.replace({'species':'[a-z]'}, '0', regex = True) #all letters from a to z in the column species are replaced with zero

#note: in above example 0 was given as integer but here it given as string
```

# method()

```python
df.replace('fire', method = 'ffill') #replaced fire with the previous value       ffill stands for forward fill
#instead of ffill we can also use 'pad'(which is default method value in repalce)

#Try it yourself: bfill method
```

# limit()

```python
#limited values are replaced
df.replace('fire', method = 'ffill', limit = 1) #1 fire value is replaced with previous value
```

# inplace()

```python
#it does not create a new data frame but modify the orginal data frame
#values to this are given in boolean
df.replace(80, 12345, inplace = True) #all NaN values are given modified with 3
df
```
