# Handeling Missing Values

```python
import pandas as pd
```

```python
df1 = pd.read_csv('D:\\Sample data\\pokemon.csv')
df1
```

# isnull()

```python
#False means there is no missing values, True means there are missing values
df1.isnull()
```

```python
#finding total null values in each column
df1.isnull().sum()
#type_2 has 402, height has 11 and speed has 10 null values while other have zero
```

```python
#finding total number of null values in csv
df1.isnull().sum().sum()
```

# notnull()

```python
#it is the reverse of isnull

#True means there is no missing values, False means there are missing values
df1.notnull()
```

```python
#finding total not null values in each column
df1.notnull().sum()
```

```python
#finding total not null values in csv
df1.notnull().sum().sum()
```

```python
#for series also we have the same functions mentioned above, apply the above function on sample series given below, and try youtself:
import numpy as np
sr = pd.Series([1,2,3,np.nan,6,7,np.NAN,9])
sr
```

# dropna()

```python
#drop all the rows(deafult) with NaN values
df1.dropna()
```

```python
#drop all the column with NaN values
df1.dropna(axis = 1) 
```

```python
#try yourself

# df1.dropna(how = 'any')             #same as df1.dropna()
# df1.dropna(axis = 1, how = 'any')   #same as df1.dropna(axis = 1)
# df1.dropna(how = 'all')             #drop only if all row or column values are null
```

```python
df1.dropna(thresh = 12) #will print row(default as axis not defined) which have atleast 12 non null values in row
```

```python
df1.dropna(thresh = 15) #will print row(default as axis not defined) which have atleast 15 non null values in row
```

```python
df1.dropna(subset = ['type_2']) #drop rows(default) with null values in cloumn mentioned
```

```python
df1.dropna(inplace = True)
df1
#dropped NaN rows an given the output it has not created any new data frame
```
